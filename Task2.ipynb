```python
# HOMEWORK
# Determine frequency dist  and CFD of any one presidentual inaugural  address
import nltk
from nltk.probability import ConditionalFreqDist
from nltk.corpus import inaugural
```

```python
text = inaugural.words(fileids='2005-Bush.txt')[:50]
fd= nltk.FreqDist(text)
fd
```

```python
cfd= ConditionalFreqDist((len(word),word) for word in text)
cfd[5]
```

```python
#Stemming Paragraphs
from nltk.stem import PorterStemmer
sp = PorterStemmer()
text = "Muskan is the best"
text = [sp.stem(token) for token in text.split(" ")]
print(" ".join(text))
```

```python
#Lemmatizer
from nltk.stem import WordNetLemmatizer
lm = WordNetLemmatizer()
lm.lemmatize('nuclei')
```

```python
lm.lemmatize('better', pos ='a') #a= adjective
```
